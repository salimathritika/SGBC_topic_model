```python
import re

def extract_code(url):
    if pd.isna(url):
        return None
    patterns = [
        r'/obo/([A-Za-z0-9\-\.]+)_',
        r'nifstd/([A-Za-z0-9_\-]+)_',
        r'http://www.ncbi.nlm.nih.gov/([^/]+)/[^/]+$',  #gene
        r'http://uri.neuinfo.org/nif/nifstd/([a-zA-Z]+)[0-9]+$',  #for sao
        r'/base/([A-Za-z0-9\-\.]+)_',  #for ilx
        r'#([A-Z]+)_',  #for cao
        r'http://www.cogpo.org/ontologies/([a-zA-Z]+)_[0-9]+$' ,
        r'http://uri.interlex.org/fsl/uris/atlases/([^/]+)/'  #for talairach
    ]
    for pattern in patterns:
        match = re.search(pattern, url)
        if match:
            return match.group(1)
    return None
```

```python
import pandas as pd
import requests
import re

# Step 1: Map Graph IDs to class codes
graph_id_map = {
    1: 'mouse_brain_atlas',
    17: 'developing_mouse_brain_atlas',
    10: 'human_brain_atlas',
    16: 'developing_human_brain_atlas',
    8: 'nonhuman_primate_brain_atlas',
    15: 'glioblastoma'
}

# Step 2: Target rows where class_code is missing & URL is from brain-map API
mask = df['class_code'].isna() & df['Class ID'].str.contains('api.brain-map.org/api/v2/data/Structure', na=False)

# Step 3: Iterate over filtered rows and extract graph_id
for idx, row in df[mask].iterrows():
    url = row['Class ID']
    match = re.search(r'/Structure/(\d+)', url)
    if match:
        structure_id = match.group(1)
        api_url = f"http://api.brain-map.org/api/v2/data/Structure/{structure_id}"
        try:
            response = requests.get(api_url, timeout=5)
            if response.status_code == 200:
                data = response.json()
                graph_id = data['msg'][0].get('graph_id')
                atlas_code = graph_id_map.get(graph_id, f"brainmap_graph_{graph_id}")
                df.at[idx, 'class_code'] = atlas_code
                print(f"Assigned '{atlas_code}' to row {idx} with graph_id {graph_id}")
            else:
                print(f"Failed to fetch {structure_id}: HTTP {response.status_code}")
        except Exception as e:
            print(f"Error fetching {structure_id}: {e}")
```

```python
import requests
import re

def get_graph_id_from_url(url):
    """
    Given a URL from api.brain-map.org Structure endpoint,
    this function returns the graph_id (integer) or None if not found.
    """
    # Extract structure ID from URL
    match = re.search(r'/Structure/(\d+)', url)
    if not match:
        return None

    structure_id = match.group(1)
    api_url = f"http://api.brain-map.org/api/v2/data/Structure/{structure_id}"

    try:
        response = requests.get(api_url, timeout=5)
        if response.status_code == 200:
            data = response.json()
            return data['msg'][0].get('graph_id')
        else:
            print(f"Failed to fetch data: HTTP {response.status_code}")
            return None
    except Exception as e:
        print(f"Error fetching graph_id: {e}")
        return None
```
